```python
def catigorize_col (df, col, labels):

    edges = [df[col].describe()['min'],
             df[col].describe()['25%'],
             df[col].describe()['50%'],
             df[col].describe()['75%'],
             df[col].describe()['max']]

    df[col] = pd.cut(df[col], edges, labels = labels, duplicates='drop')
    return df
```

```python
labels = ['not_popular', 'below_avg', 'average', 'popular']
catigorize_col(df, 'vote_average', labels)

df['vote_average'].unique()
```

```python
df.dropna(inplace = True)
df.isnull().sum()
```

```python

def categorize_col(df, col, labels):
    edges = [
        df[col].min(),
        df[col].quantile(0.33),
        df[col].quantile(0.66),
        df[col].max()
    ]
    df[col + '_category'] = pd.cut(
        df[col],
        bins=edges,
        labels=labels,
        include_lowest=True
    )
    return df
labels = ['low', 'average', 'high']
df = categorize_col(df, 'profit', labels)
print(df)
print(df['profit_category'].unique())
```

Q2: What are the most filmed genres in this whole dataset?

```python
top_3_genres = df['genres'].value_counts().nlargest(3)
plt.figure(figsize=(10, 5))
ax = top_3_genres.plot( kind="bar", color=["#FF9999", "#66B2FF", "#99FF99"],  edgecolor="black", width=0.6)
plt.title("Top 3 Most Filmed Genres", fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Genres", fontsize=14, labelpad=10)
plt.ylabel("Count", fontsize=14, labelpad=10)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=12, padding=3)
plt.tight_layout()
plt.show()
print("Top 3 Genres and Counts:")
print(top_3_genres)
```
